# test, train spliting

```python
X=df.drop(['y_yes'],axis=1)
Y=df['y_yes']
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size=0.2,random_state=0)
```

# QN-3    supervised learning models!!

## LinearRegression

```python
model = LinearRegression()
model.fit(train_x,train_y)
```

```python
model.predict(test_x)
```

```python
mean_squared_error(test_y, model.predict(test_x), squared=False)
```

## AdaBoosting

```python
regr = AdaBoostRegressor(random_state=0, n_estimators=50)
regr.fit(train_x,train_y)
```

```python
mean_squared_error(test_y, regr.predict(test_x), squared=False)
```

```python
def AdaBoostReg(train_x,train_y,test_x):
    modelabr = AdaBoostClassifier(random_state=42)
    modelabr.fit(train_x,train_y)
    return modelabr.predict(test_x)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(train_x,train_y)
y_head=nb.predict(test_x)
```

```python
nb.score(test_x,test_y)
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
k=8
knn = KNeighborsClassifier(n_neighbors = k)
knn.fit(train_x,train_y)
y_head=knn.predict(test_x)
```

```python
knn.score(test_x,test_y)
```

## SVM

```python
from sklearn.svm import SVC # "Support vector classifier"  
classifier = SVC(kernel='linear', random_state=0)  
classifier.fit(train_x, train_y)  
```

```python
y_pred= classifier.predict(test_x)
```

```python
classifier.score(test_x,test_y)
```
